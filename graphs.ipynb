```python
import matplotlib.pyplot as plt 
import json

with open("filtered_data.json") as f:
        data_list = json.load(f)

def get_info(movie: int,info: str):
        """
        :param movie: list id for movies from data_list
        :param info: the type of info you want from data_list. \nPossible inputs: genre_ids, id, original_language,title,vote_avrage,vote_count.
        :return: returns the info from the movie selected
        """
        return data_list[movie][info]

def get_all_info(info: list) -> list:
    wanted_info_list = []
    """
    :param info: list of info you want from data_list
    :return: returns the info for all data in data_list as a list, 
    """
    for i in range(len(data_list)):
        temp = []
        for j in info:
            temp.append(get_info(i,j))
        wanted_info_list.append(temp)
    return wanted_info_list

def make_graph(info: list):
    """
    """
    x = []
    y = []
    for i in range(len(info)):
        x.append(info[i][0])
    for i in range(len(info)):
        y.append(info[i][1])
    plt.plot(x, y)
    plt.title("Ratings")
    plt.plot()

print(get_all_info(["title","vote_average"]))
print(make_graph(get_all_info(["title","vote_average"])))
```
